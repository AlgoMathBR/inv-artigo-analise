```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Analisar o Dataset

```python
# Ler os dataset
df = pd.read_csv('./build_dataset/dataset.csv.txt', encoding='utf-8')
df_1 = pd.read_csv('./build_dataset/freedom.txt', encoding='utf-8')
```

```python
# Analisa em separado o Indice de Liberdade Economica com o GDP per Capita
index_max_year = df.groupby('Entity')['Year'].idxmax()
df_2 = df.loc[index_max_year]

x = df[(df['GDP per capita'] != 0.0)]
index_max_year = x.groupby('Entity')['Year'].idxmax()
x1 = x.loc[index_max_year]

y = pd.merge(x, df_1[['Entity', 'Score']], on='Entity', how='left')
index_max_year2 = y.groupby('Entity')['Year'].idxmax()
y1 = y.loc[index_max_year2]

df_sem_nan = y1.dropna()

correlacao_ = df_sem_nan['GDP per capita'].corr(df_sem_nan['Score'])

plt.figure(figsize=(8, 6))
plt.scatter(df_sem_nan['GDP per capita'], df_sem_nan['Score'], alpha=0.5)
plt.title('Correlação entre {} e {}'.format('GDP per capita', 'Index of Economic Freedom'))
plt.xlabel('GDP per capita')
plt.ylabel('Index of Economic Freedom')
plt.grid(True)
plt.show()

print('Correlação entre {} e {}:'.format('GDP per capita', 'Index of Economic Freedom'), correlacao_)
```

```python
# Analisa em separado o Indice de Liberdade Economica com o GDP
index_max_year = df.groupby('Entity')['Year'].idxmax()
df_2 = df.loc[index_max_year]

x = df[(df['GDP'] != 0.0)]
index_max_year = x.groupby('Entity')['Year'].idxmax()
x1 = x.loc[index_max_year]

y = pd.merge(x, df_1[['Entity', 'Score']], on='Entity', how='left')
index_max_year2 = y.groupby('Entity')['Year'].idxmax()
y1 = y.loc[index_max_year2]

df_sem_nan = y1.dropna()

correlacao_ = df_sem_nan['GDP'].corr(df_sem_nan['Score'])

plt.figure(figsize=(8, 6))
plt.scatter(df_sem_nan['GDP'], df_sem_nan['Score'], alpha=0.5)
plt.title('Correlação entre {} e {}'.format('GDP', 'Index of Economic Freedom'))
plt.xlabel('GDP')
plt.ylabel('Index of Economic Freedom')
plt.grid(True)
plt.show()

print('Correlação entre {} e {}:'.format('GDP', 'Index of Economic Freedom'), correlacao_)
```

```python
def correlacao_entre_variaveis(var1, var2, df):

    df2 = df[(df[var1] != 0.0) & (df[var2] != 0.0)]
    df3 = df2[['Entity', 'Year', var1, var2]]

    index_max_year = df3.groupby('Entity')['Year'].idxmax()
    df4 = df3.loc[index_max_year]
    correlacao = df4[var1].corr(df4[var2])

    plt.figure(figsize=(8, 6))
    plt.scatter(df4[var1], df4[var2], alpha=0.5)
    plt.title('Correlação entre {} e {}'.format(var1, var2))
    plt.xlabel(var1)
    plt.ylabel(var2)
    plt.grid(True)
    plt.show()

    print('Correlação entre {} e {}:'.format(var1, var2), correlacao)
```

```python
correlacao_entre_variaveis('GDP per capita','Life Expectancy',df)
correlacao_entre_variaveis('GDP per capita','IDH',df)
correlacao_entre_variaveis('GDP per capita','Average annual working hours per worker',df)
correlacao_entre_variaveis('GDP per capita','MtCO2e',df)
correlacao_entre_variaveis('GDP per capita','Inequality in income',df)
correlacao_entre_variaveis('GDP per capita','Material footprint per capita',df)
correlacao_entre_variaveis('GDP per capita','Poverty Index',df)
correlacao_entre_variaveis('GDP per capita','Population Total',df)
correlacao_entre_variaveis('GDP per capita','business',df)

correlacao_entre_variaveis('GDP','Life Expectancy',df)
correlacao_entre_variaveis('GDP','IDH',df)
correlacao_entre_variaveis('GDP','Average annual working hours per worker',df)
correlacao_entre_variaveis('GDP','MtCO2e',df)
correlacao_entre_variaveis('GDP','Inequality in income',df)
correlacao_entre_variaveis('GDP','Material footprint per capita',df)
correlacao_entre_variaveis('GDP','Poverty Index',df)
correlacao_entre_variaveis('GDP','Population Total',df)
correlacao_entre_variaveis('GDP','business',df)

correlacao_entre_variaveis('GDP','GDP per capita',df)
```
